# Chapter 7 - Derivatives

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x
x = np.array(range(5))
y = f(x)

plt.plot(x, y)
plt.show()
```

## The Slope

$$
\cfrac{Change\ in\ y}{Change \ in\ x} = \cfrac{\Delta y}{\Delta x}
$$
![how to calculate the slope](images/07-1.png)


```python
print((y[1]-y[0]) / (x[1]-x[0]))
```

What about calculating the slope for non-linear functions such as
$$
f(x)=2x^2
$$

```python
def g(x):
    return 2*x**2
```

```python
y2 = g(x)

plt.plot(x, y2)
plt.show()
```

Depending on the points we take the slope will change:

![how to calculate the slope](images/07-2.png)

Chetk [this video](https://youtu.be/aNaFf6InhZY)

```python
print((y2[1]-y2[0]) / (x[1]-x[0]))
print((y2[2]-y2[1]) / (x[2]-x[1]))
print((y2[3]-y2[2]) / (x[3]-x[2]))
```

To give an immediate example, we can approximate a derivative of the function at *x* by using this point and another one also taken at x, but with a minimal delta added to it, such as *0.0001*.

This number is a common choice as it does not introduce too large an error (when estimating the derivative)

```python
p2_delta = 0.0001

x1 = 1
x2 = x1 + p2_delta  # add delta

yg1 = g(x1)  # result at the derivation point
yg2 = g(x2)  # result at the other, close point

approximate_derivative = (yg2-yg1)/(x2-x1)
print(approximate_derivative)
```
